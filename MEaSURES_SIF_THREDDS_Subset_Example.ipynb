# Spatial and Temporal Subsetting of Gridded MEaSURES SIF Data  

This tutorial demonstrates two simple scenarios of subsetting gridded Solar-Induced Chlorophyll Fluorescence Earth System Data Record (SIF-ESDR) data available from the ORNL DAAC's Thematic Real-time Environmental Distributed Data Services (THREDDS) Data Server (TDS) using Python.  
- Scenario 1: Spatial subsetting within a region at particular time points.  
- Scenario 2: Time series subsetting at a particular point location.  

The aim of these scenarios is to investigate how estimates of SIF in the Great Smoky Mountains National Park (GSMNP) were affected by the [2016 Great Smoky Mountains wildfires](https://en.wikipedia.org/wiki/2016_Great_Smoky_Mountains_wildfires).  

**Datasets Used in this Tutorial**  

Two datasets are used in the tutorial to demonstrate the “interoperability” of ORNL DAAC data products: it is easy to use different SIF data products in the same analysis workflow while making minimal changes.  

- [High Resolution Global Contiguous SIF Estimates from OCO-2 SIF and MODIS, Version 2](https://doi.org/10.3334/ORNLDAAC/1863) (hereafter, Dataset 1863)  
- [Global High-Resolution Estimates of SIF from Fused SCIAMACHY and GOME-2, 2002-2018](https://doi.org/10.3334/ORNLDAAC/1864) (hereafter, Dataset 1864)

## 2. Prepare the Data.

The tutorial will begin with Dataset 1863. You can run the code using Dataset 1864 later to compare how the results change using different data.

#### !!!!NEED EXPLAINATIONS OF SIF AND WHAT THE DATA MEAN

```python
# URL of Dataset 1863 in the ORNL DAAC THREDDS Data Server
tds_url = 'https://thredds.daac.ornl.gov/thredds/catalog/ornldaac/1863-agg/catalog.xml'

# Name of the data variable containing the SIF measurements for Dataset 1863
vname = 'sif_ann'

# Adjust the SIF levels to match SIF data values from different datasets (for visualization purposes)
sif_levels = np.linspace(0.0, 0.7, 151)
```

## 3. Take a look at the study area and wildfire event.

The GSMNP is an American national park in the southeastern United States, with parts in both Tennessee and North Carolina. A complex of wildfires began in late November 2016, and the towns most impacted were Pigeon Forge and Gatlinburg, both near GSMNP. Fire claimed 14 lives, injured 190, and is one of the largest natural disasters in the history of Tennessee.

In this tutorial, we will examine a map of SIF estimates around the GSMNP in the summer following the wildfire event and how the measurements at the center of the wildfires changed.

The code below will load and display a shapefile representing the boundary of GSMNP, extract the bounding box (*bbox*) of the shapefile, and also define the latitude and longitude coordinates of the wildfire's center (denoted by the red dot).

```python
# Define the latitude and longitude coordinates of the wildfires center point
fire_center_lat = 35.63
fire_center_lon = -83.48

# Load the shapefile
gsmnp_bnd_4326 = gpd.read_file('bnds/GRSM_BOUNDARY_POLYGON_4326.shp')

# Extract bbox of the shapefile
gsmnp_bbox = gsmnp_bnd_4326.bounds

# Increase the north bound by about 5 km
gsmnp_bbox.maxy = gsmnp_bbox.maxy + 0.05

# Display the shapefile and the wildfire center
gsmnp_bnd_3857 = gsmnp_bnd_4326.to_crs(epsg=3857)
ax = gsmnp_bnd_3857.plot(figsize=(12,6), alpha=0.3, edgecolor='red')

# !!!!NEED EXPLAINATIONS
# Construct a pandas DataFrame object with three fields: longtiude, latitude, and place_name
gsmnp_bnd_df = pd.DataFrame({'longitude':[fire_center_lon], 'latitude':[fire_center_lat], 'place_name':['2016 Great Smoky Mountains Wildfires']})

# !!!!NEED EXPLAINATIONS FOR DIFFERENCE BETWEEN 4326 & 3857
# Convert the pandas DataFrame object into a GeoPandas GeoDataFrame object and set Geographic Lat/Lon (EPSG:4326) as its CRS
fire_place_4326 = gpd.points_from_xy(gsmnp_bnd_df.longitude, gsmnp_bnd_df.latitude, crs="EPSG:4326")
gsmnp_bnd_geo_df_4326 = gpd.GeoDataFrame(gsmnp_bnd_df[['latitude','longitude','place_name']], geometry=fire_place_4326)

# Leverage GeoPandas' spatial operation to reproject the GeoDataFrame object from Geographic Lat/Lon (EPSG:4326) onto Web Mercator (EPSG:3857)
# This reprojection step is purely for visualization purpose, in order to overlay with the OpenStreetMap base layer (see below)
fire_place_3857 = fire_place_4326.to_crs(epsg=3857)
gsmnp_bnd_geo_df_3857 = gpd.GeoDataFrame(gsmnp_bnd_df[['latitude','longitude','place_name']], geometry=fire_place_3857)
gsmnp_bnd_geo_df_3857.plot(column='place_name', ax=ax, cmap=ListedColormap('red'), markersize=100, legend=True)

# !!!!NEED EXPLAINATIONS
# Add the OpenStreetMap map as a base layer. 
# Note that the OpenStreetMap layer only supports the Web Mercator (EPSG:3857) projection
ctx.add_basemap(ax)
```

## 4. Obtain data from the ORNL DAAC's THREDDS Data Server.

### 4.1 Open a remote SIF dataset.

```python
# !!!!NEED EXPLAINATIONS
# Use Siphon to connect to the ORNL DAAC THREDDS Data Server catalog
tds = catalog.TDSCatalog(tds_url) 
# List the datasets available in this catalog
datasets = list(tds.datasets)
print(datasets)
```

### 4.2. List the types of service APIs that can access the dataset.

The ORNL DAAC TDS supports data access using different service APIs, like OPeNDAP, NetCDF Subset Service, and OGC Web Coverage Service. We will use the NetCDF Subset Service (NCSS) API in this tutorial.

```python
# !!!!NEED EXPLAINATIONS
# List data access options for the first dataset available from the catalog
endpts = list(tds.datasets.values())
for key, value in endpts[0].access_urls.items():
    print(key, value)
```

### 4.3. Select the NCSS API.

#### !!!!NEED EXPLAINATIONS FOR PRINTING THE COMMANDS

```python
# !!!!NEED EXPLAINATIONS
# Pick the NetCDF Subset Service (NCSS) access option
ncss = endpts[0].subset()

# Print the data variables in the dataset
print("Variables: " + str(ncss.metadata.variables))
# Print the full temporal range of the dataset
print("Temporal Extent: " + str(ncss.metadata.time_span))
# Print the full spatial extent (or bbox) of the dataset
print("Spatial Extent: " + str(ncss.metadata.lat_lon_box))
```

## 5. Spatially subset the SIF data for the GSMNP in Summer 2016 and 2017.

### 5.1. Build a NCSS spatial subset request.

```python
# Create an empty NCSS query object
spatial_query = ncss.query()
# Subset the SIF data variable in netCDF version 4 format
spatial_query.variables(vname).accept('netcdf4')
# Subset within the bbox of GSMNP
spatial_query.lonlat_box(north=gsmnp_bbox.maxy, south=gsmnp_bbox.miny, east=gsmnp_bbox.maxx, west=gsmnp_bbox.minx)
```

### 5.2. Retrieve data for Summer 2016.

```python
# Subset data at a time step closest to 2016-06-30
spatial_query.time(datetime.datetime(2016,6,30,0,0))
# Print the query string
print("Spatial Query: " + str(spatial_query))
# Perform the subset
spatial_data_2016_response = ncss.get_data(spatial_query)
```

### 5.3. Visualize the data for Summer 2016.

```python
# !!!!NEED EXPLAINATIONS
# !!!!ADD LEGEND TITLE
# This codes section plots three data layers on a map in the Geographic Lat/Lon CRS (EPSG:4326)
#   1. SIF data for Summar 2016 in our study area, represented by 'sif_spatial_data'
#   2. Boundary polygon for the Great Smoky Mountain National Park (GSMNP), represented by 'gsmnp_bnd_4326'
#   3. Center point of the 2016 GSMNP wild fire event, represented by (spatial_lat, spatial_lon)
spatial_lon = spatial_data_2016_response.variables['lon'][:].squeeze()
spatial_lat = spatial_data_2016_response.variables['lat'][:].squeeze()
sif_spatial_data = spatial_data_2016_response.variables[vname][:].squeeze()
spatial_lon_2d, spatial_lat_2d = np.meshgrid(spatial_lon, spatial_lat)
fig_spatial_2016 = plt.figure(figsize=(16, 8))
ax_spatial_2016 = plt.axes(projection=ccrs.PlateCarree())
ax_spatial_2016.set_extent([gsmnp_bbox.maxx, gsmnp_bbox.minx, gsmnp_bbox.miny, gsmnp_bbox.maxy])
ax_spatial_2016.set_title("SIF Estimates for Summer 2016 in the Great Smoky Mountain National Park (units: " + spatial_data_2016_response[vname].units + ")" , fontsize=16)
cf_spatial_2016 = ax_spatial_2016.contourf(spatial_lon_2d, spatial_lat_2d, sif_spatial_data, 200, levels=sif_levels, transform=ccrs.PlateCarree(), zorder=0, cmap='RdYlGn')
plt.colorbar(cf_spatial_2016, ax=ax_spatial_2016, fraction=0.032)
ax_spatial_2016.plot(spatial_lon_2d.flatten(), spatial_lat_2d.flatten(), marker='o', color='black', markersize=2,
        alpha=0.3, transform=ccrs.Geodetic(), zorder=2, linestyle='none')
gsmnp_bnd_4326.plot(ax=ax_spatial_2016, alpha=1, facecolor="none", edgecolor='red')
gsmnp_bnd_geo_df_4326.plot(column='place_name', ax=ax_spatial_2016, cmap=ListedColormap('red'), markersize=100, legend=True)
```

### 5.4. Update the NCSS spatial subset request to retrieve data for Summer 2017.

```python
# Subset data at a time step closest to 2017-06-30
spatial_query.time(datetime.datetime(2017,6,30,0,0))
# Print the query string
print("Spatial Query: " + str(spatial_query))
# Perform the subset
spatial_data_2017_response = ncss.get_data(spatial_query)
```

### 5.5. Visualize the returned data for Summer 2017.

```python
# !!!!NEED EXPLAINATIONS
# !!!!ADD LEGEND TITLE
spatial_lon = spatial_data_2017_response.variables['lon'][:].squeeze()
spatial_lat = spatial_data_2017_response.variables['lat'][:].squeeze()
sif_spatial_data = spatial_data_2017_response.variables[vname][:].squeeze()
spatial_lon_2d, spatial_lat_2d = np.meshgrid(spatial_lon, spatial_lat)
fig_spatial_2017 = plt.figure(figsize=(16, 8))
ax_spatial_2017 = plt.axes(projection=ccrs.PlateCarree())
ax_spatial_2017.set_extent([gsmnp_bbox.maxx, gsmnp_bbox.minx, gsmnp_bbox.miny, gsmnp_bbox.maxy])
ax_spatial_2017.set_title("SIF Estimates for Summer 2017 in the Great Smoky Mountain National Park (units: " + spatial_data_2017_response[vname].units + ")" , fontsize=16)
cf_spatial_2017 = ax_spatial_2017.contourf(spatial_lon_2d, spatial_lat_2d, sif_spatial_data, 200, levels=sif_levels, transform=ccrs.PlateCarree(), zorder=0, cmap='RdYlGn')
plt.colorbar(cf_spatial_2017, ax=ax_spatial_2017, fraction=0.032)
ax_spatial_2017.plot(spatial_lon_2d.flatten(), spatial_lat_2d.flatten(), marker='o', color='black', markersize=2,
        alpha=0.3, transform=ccrs.Geodetic(), zorder=2, linestyle='none')
gsmnp_bnd_4326.plot(ax=ax_spatial_2017, alpha=1, facecolor="none", edgecolor='red')
gsmnp_bnd_geo_df_4326.plot(column='place_name', ax=ax_spatial_2017, cmap=ListedColormap('red'), markersize=100, legend=True)
```

### 5.5. Compare the Summer 2016 map to the Summer 2017 map.
Comparing the two maps above, there are lower (redder) SIF values near the wildefire's center in Summer 2017 than in Summer 2016. This implies that the wildfires had a negative impact on SIF in the GSMNP area.

## 6. Temporally subset the SIF data at the wildfires center for the time period 2015-2019.

### 6.1. Build a NCSS temporal subset request and retrieve the data.

```python
# Create an empty NCSS query object
temporal_query = ncss.query()
# Subset the SIF data variable in netCDF version 4 format
temporal_query.variables(vname).accept('netcdf')
# Subset the SIF data to the wildfire center
temporal_query.lonlat_point(fire_center_lon, fire_center_lat)
# Subset data at a time step period of 2015-01-1 to 2019-12-31
temporal_query.time_range(datetime.datetime(2015,1,1,0,0), datetime.datetime(2019,12,31,0,0))
# Print the query string
print("Temporal Query: " + str(temporal_query))
# Perform the subset
temporal_data_response = ncss.get_data(temporal_query)
```

### 6.2. Prepare the data and time values for visualization.

```python
# !!!!NEED EXPLAINATIONS
# Retrieve the raw sif data values from the NCSS response
sif_temporal_data = temporal_data_response.variables[vname]
print(sif_temporal_data[:].squeeze())

# Convert time values to human-readable format
time = temporal_data_response.variables['time']
time_vals = num2pydate(time[:].squeeze(), time.units)
#print(time_vals)
```

### 6.3. Make a SIF time series plot.

```python
# !!!!NEED EXPLAINATIONS
# Plot the SIF time series data
fig_temporal, ax_temporal = plt.subplots(1, 1, figsize=(12, 5))
ax_temporal.plot(time_vals, sif_temporal_data[:].squeeze(), 'r', linewidth=2)
ax_temporal.set_title('{} ({})'.format(sif_temporal_data.long_name, sif_temporal_data.units))
ax_temporal.set_xlabel('Time (UTC)')
ax_temporal.grid(True)
```

### 6.4. Examine and describe the time series plot.
In the time series plot, the wildefire center (at ~5 km grid cell area) the annual peak values were typically above 0.6 prior to 2016. In summer 2017, the year after the 2016 wildfires, the peak value decreased to ~0.52. Also, there is an increasing trend SIF peak values from 2017 to 2019, which means vegetation was recovering. One might conclude that the 2016 wildfires had a negative impact on SIF at the fire center point.
